```
import tensorflow as tf
from sklearn.datasets import load_iris
import numpy as np
from keras.utils import to_categorical
```

```
x = load_iris().data
y = load_iris().target

x[:10], y[:10]
```

```
np.unique(y)
```

```
# [batch_size, 4]
input_features = tf.placeholder(dtype=tf.float32, shape=[None, x.shape[1]])

# [batch_size, 3]
input_labels = tf.placeholder(dtype=tf.float32, shape=[None, 3])

# Neural network's first layer
weights = tf.Variable(tf.truncated_normal(shape=[x.shape[1], 5], stddev=0.1))
biases = tf.Variable(tf.zeros(shape=[5]))

# Neural network's output layer
weights2 = tf.Variable(tf.truncated_normal(shape=[5, 3], stddev=0.1))
biases2 = tf.Variable(tf.zeros(shape=[3]))

# Neural network model
neural_model1 = tf.nn.relu(tf.matmul(input_features, weights) + biases)
neural_model2 = tf.nn.softmax(tf.matmul(neural_model1, weights2) + biases2)

# average (-y * log(w * x + b))
loss = tf.reduce_mean(-tf.reduce_sum(input_labels * tf.log(neural_model2), reduction_indices=[1]))

train_op = tf.train.GradientDescentOptimizer(learning_rate=0.001).minimize(loss)
```

```
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```
def next_batch(batch_size, features, labels):
    indices = np.arange(start=0, stop=features.shape[0])
    np.random.shuffle(indices)
    indices = indices[:batch_size]
    
    return features[indices], labels[indices]
```

```
batch_size = 16

for epoch in range(10):
    for index in range(int(x.shape[0] / batch_size)):
        minibatch_x, minibatch_y = next_batch(batch_size=batch_size, features=x, labels=y)
        
        minibatch_y = to_categorical(minibatch_y)
        _, train_loss = sess.run([train_op, loss], feed_dict={
                        input_features: minibatch_x,
                        input_labels: minibatch_y
                    })
    print('Epoch {}, loss : {}'.format(epoch, train_loss))
```
